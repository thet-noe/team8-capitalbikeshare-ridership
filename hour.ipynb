```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file = '/Users/omaradriannecarpiobapora/Desktop/CS-133/team8-capitalbikeshare-ridership/data/hour.csv'
hour = pd.read_csv(csv_file)
```

```python
hour.max()
```

```python
hour = hour.astype({
    'instant':'int16',
    'season':'int16',
    'yr':'int16',
    'mnth':'int16',
    'hr':'int16',
    'holiday':'int16',
    'weekday':'int16',
    'workingday':'int16',
    'weathersit':'int16',
    'casual':'int16',
    'registered':'int16',
    'count':'int16',
})
```

```python
g=sns.pairplot(
    data=hour,
    kind='reg',
    plot_kws={'scatter_kws':{'color':'lightgrey', 's':5}},
    height=1.5
)
g.fig.suptitle('Pairwise Relationships in Capital Bikeshare Dataset', y=1.05);
```

```python
# Small-multiples plot: relplot()
g=sns.relplot(
    data=hour,
    x='hr',
    y='count',
    col='season',
    col_wrap=2,
    hue='temp',
    palette='GnBu'
)
g.fig.suptitle('Ridership Count vs Hour by Season', y=1.03);
```

```python
# Regression plot (lmplot)
g=sns.lmplot(
    data=hour,
    x='temp',
    y='count',
    scatter_kws={'color':'pink', 'alpha':0.4}
)
g.fig.suptitle('Ridership Count vs Temperature', y=1.03);
```

```python
def to_text_season(season):
    if season == 1:
        return 'Winter'
    elif season == 2:
        return 'Spring'
    elif season == 3:
        return 'Summer'
    elif season == 4:
        return 'Fall'
        

hour['tseason'] = hour['season'].map(to_text_season)
display(hour['tseason'].head())
display(hour['tseason'].value_counts())
```

```python
# Categorical plot: Box Plot
g=sns.catplot(
    data=hour,
    kind='box',
    x='count',
    y='tseason',
    showmeans=True
)
g.fig.suptitle('Ridership Count Distribution by Season', y=1.03);
```
